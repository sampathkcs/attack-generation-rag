```python
import requests
from bs4 import BeautifulSoup
import json
import re
import pandas as pd
```

```python
# Make a request
page = requests.get(
    "https://attack.mitre.org/tactics/ics/")
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
soup
```

```python
# Extract body of page
page_body = soup.body
```

```python
table = soup.find('table', attrs={'class':'table table-bordered table-alternate mt-2'})
table_rows = table.find_all('tr')
```

```python
data = []
for row in table.find_all('tr'):
    row_data = []
    for cell in row.find_all('td'):
        row_data.append(cell.text)
    data.append(row_data)
```

```python
data
```

```python
data_list = []
```

```python
for list in data:
    if len(list)==3:
        data_list.append({  
                            "ID" : list[0].replace("\n","").strip(),
                            "Name" : list[1].replace("\n","").strip(),
                            "Description" : list[2].replace("\n","").strip()
                        })
```
